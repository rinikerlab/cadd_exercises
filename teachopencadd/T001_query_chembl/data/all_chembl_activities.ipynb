# Query local ChEMBL DB for speed

Get your local copy from http://doi.org/10.6019/CHEMBL.database.27. You need `chembl_27_sqlite.tar.gz` for this notebook. Extract it and point `DATABASE` below to the `chembl_27.db` file.

```python
DATABASE = "chembl_27/chembl_27_sqlite/chembl_27.db"
conn = sql.connect(DATABASE, isolation_level=None)
db = conn.cursor()
```

```python
q = "SELECT activities.activity_id FROM activities"
result = db.execute(q)
```

```python
arr = np.array(list(result), dtype=int).reshape((-1,))
```

```python
np.savez_compressed("chembl27_activities", activities=arr, allow_pickle=False)
```

Load compressed `*.npz` file like this. Resulting `activities` is a `set` with fast membership tests!

```python
import numpy as np
```

```python
with np.load("chembl27_activities.npz") as f:
    activities = set(f["activities"])
```
